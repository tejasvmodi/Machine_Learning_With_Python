```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
```

```python
sns.lineplot(x="total_bill",y="tip",data=tips)
plt.title('total_bill vs tip')
```

```python
titenic=sns.load_dataset("titanic")
```

```python
titenic.head()
```

```python
sns.scatterplot(x="fare",y="age",data=titenic)
```

```python
sns.lineplot(x="fare",y="age",data=titenic)
sns.set_style("dark")
```

```python
sns.countplot(x='age', data=titenic) 
plt.show()
```

```python
sns.histplot(x='age', y='fare', data=titenic) 
```

```python
data = [44, 45, 40, 41, 39] 
keys = ['Class 1', 'Class 2', 'CLass 3', 'Class 4', 'Class 5'] 
palette_color = sns.color_palette('bright') 
plt.pie(data, labels=keys, colors=palette_color) 
plt.show()
```

```python
sns.barplot(x="age",y="fare",data=titenic,label="total")
```

```python
sns.boxplot(data=titenic, x="age", y="fare", hue="alive")
```

```python
sns.barplot(titenic, x="fare", y="age", hue="alive")
```
