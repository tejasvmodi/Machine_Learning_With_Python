```python
#sum the all column 
pokemon['Total']=pokemon['against_dark']+pokemon['against_steel']+pokemon['against_fairy']+pokemon['against_ghost']
print(pokemon)
```

```python
#create a new column name avg in csv file 
pokemon['Avg']=pokemon['Total']/4
print(pokemon)
```

```python
#create a new csv file 
pokemon.to_csv('pokemon.csv')
```
