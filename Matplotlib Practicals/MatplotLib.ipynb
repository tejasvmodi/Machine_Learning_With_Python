```python
import pandas as pd
import numpy as np
#the import file of matplot
import matplotlib.pyplot as plt
```

```python
x=np.array([5,8,15,69,80])
y=np.array([3,9,25,37,90])
#plot the graph with single cordinate 

plt.plot(x,y)
```

```python
x1=np.array([4,5,88,73])
y1=np.array([9,12,44,56])
# plt.plot(x1,y1)
# plt.plot(x,y)
#plot the 2 graph cordinate in one graph .......
plt.plot(x,y,x1,y1)
```

```python
#print the dots with the using the o ......
plt.plot(x,y,'o')
```

```python
#when cordinate is matched tht time the dot is printed 
plt.plot(x,y,marker='d')
```

```python
#the markers(star size ) is increased 
plt.plot(x,y,marker='*',ms=10)
```

```python
#add color in marker 
plt.plot(x,y,marker='d',mec='r')
```

```python
#fill the color inside marker 
plt.plot(x,y,marker='d',mfc='y')
```

```python
#fill the color ,border and size of the marker 
plt.plot(x,y,marker='*',mfc='y',mec='r',ms=20)
```

```python
#the graph with the linewidth 
plt.plot(x,y,linestyle='dashdot',color='g')
plt.plot(x1,y1,linestyle='dashdot',color='m',linewidth=3)
```

```python
#the chart with graph title and label...
plt.title("Demo chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,linestyle='dashdot',color='g')
plt.plot(x1,y1,linestyle='dashdot',color='m',linewidth=1)
```

```python
#change the font size of the title and the location of the title 
font={'size':15,'color':'b'}
plt.title("Demo chart",fontdict=font,loc='right')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,linestyle='dashdot',color='g')
plt.plot(x1,y1,linestyle='dashdot',color='m',linewidth=1)
```

```python
#change the font size of the title and the location of the title 
font={'size':15,'color':'b'}
plt.title("Demo chart",fontdict=font,loc='right')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,linestyle='dashdot',color='g')
plt.plot(x1,y1,linestyle='dashdot',color='m',linewidth=1)

#grid line in your graph with respect to x and y 
plt.grid()

#grid line in your graph with respect to x  
plt.grid(axis='y')

#grid line in your graph with respect to  y  and also the add the linestyle in dotted line . and add color for the grid  

plt.grid(axis='x',linestyle=':',color="b")
```

```python
#plt.subpplot(row,column,columnindex)
plt.subplot(1,2,1)
plt.title("p1")
plt.plot(x,y)
plt.subplot(1,2,2)
plt.title("p2")
plt.plot(x1,y1)
plt.suptitle("Tejasv")
```

```python
#read a data from the csv file 
pokemon=pd.read_csv('studentmark.csv')
print(pokemon)
```

```python
#fetch the record from the csv file 
print(pokemon.loc[3])
```

```python
#sum the all column 
pokemon['Total']=pokemon['WS']+pokemon['CS']+pokemon['MLP']+pokemon['FLUTTER']+pokemon['CCN']+pokemon['LARAVEL']
print(pokemon)
```

```python
#create a new column name avg in csv file 
pokemon['Avg']=pokemon['Total']/6
print(pokemon)
```

```python
#create a new csv file 
pokemon.to_csv('studentdatabasem.csv')
```
