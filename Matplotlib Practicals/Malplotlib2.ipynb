```python
import pandas as pd
import numpy as np
#the import file of matplot
import matplotlib.pyplot as plt
```

```python
x=np.array([5,8,15,69,80])
y=np.array([3,9,25,37,90])
#plot the graph with single cordinate 

x1=np.array([4,5,88,73])
y1=np.array([9,12,44,56])

plt.plot(x,y)
```

```python
#plotting the scatter graph using the matplotlib 
plt.scatter(x,y,s=40)
plt.show
```

```python
plt.scatter(x,y,color='r',s=20,marker="*")#s=size,use marker for athoer shape
plt.scatter(x1,y1,s=40,alpha=0.3,marker="X") #alpha use for the opacity for the color 
plt.show()
```

```python
plt.bar(x,y)
plt.show()#for vertical 
```

```python
plt.barh(x,y,color='m',height=4)
plt.show()#for horizontal 
```

```python
plt.bar(x,y,color='m',width=4)
plt.show()
```

```python
plt.hist(x,color='g') #is the specific column 
```

```python
#pie chart ....
track_stud=np.array([20,40,50,10,15])
track_name=np.array(["Web","Data","Mobile","CyberSecurity","IOT"])
ex=[0.2,0,0,0,0.2]
c=['r','Pink','g','m','b']
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True,colors=c)#explode for spacing , add shadow ,add color
plt.legend(title="TrackSelection",loc='upper left')
plt.show()
```
